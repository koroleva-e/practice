```python
# –í—ã—Ç–∞—â–∏–º –¥–∞–Ω–Ω—ã–µ –ø–æ –æ—Å—Ç–∞–ª—å–Ω—ã–º —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º (2015-16 –≥–≥) –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ 
# PS3 –Ω–µ –±—É–¥–µ–º —Å–º–æ—Ç—Ä–µ—Ç—å, —Ç–∫ PS4 - –Ω–æ–≤–µ–µ –µ—ë
data_xone = data_selection[data_selection['platform']=='XOne']
data_3ds = data_selection[data_selection['platform']=='3DS']
data_wiiu = data_selection[data_selection['platform']=='WiiU']
```

```python
# 1 –ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è –ø–æ –æ—Ç–∑—ã–≤–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (2015-16 –≥–≥) –¥–ª—è XOne
data_xone.plot(x='user_score', y='general_sales', kind='scatter', alpha=0.3, figsize=(8,8), grid=True)
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (XOne)')
plt.show()
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è XOne
xone_user_cor = data_xone['user_score'].corr(data_xone['general_sales'])
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (XOne): ', xone_user_cor)


# 2 –ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è –ø–æ –æ—Ç–∑—ã–≤–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (2015-16 –≥–≥) –¥–ª—è 3DS
data_3ds.plot(x='user_score', y='general_sales', kind='scatter', alpha=0.3, figsize=(8,8), grid=True)
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (3DS)')
plt.show()
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è 3DS
user_cor_3ds = data_3ds['user_score'].corr(data_3ds['general_sales'])
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (3DS): ', user_cor_3ds)


# 3 –ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è –ø–æ –æ—Ç–∑—ã–≤–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (2015-16 –≥–≥) –¥–ª—è WiiU
data_wiiu.plot(x='user_score', y='general_sales', kind='scatter', alpha=0.3, figsize=(8,8), grid=True)
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (WiiU)')
plt.show()
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è WiiU
wiiu_user_cor = data_wiiu['user_score'].corr(data_wiiu['general_sales'])
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (WiiU): ', wiiu_user_cor)
```

```python
# 1 –ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è –ø–æ –æ—Ç–∑—ã–≤–∞–º –∫—Ä–∏—Ç–∏–∫–æ–≤ (2015-16 –≥–≥) –¥–ª—è XOne
data_xone.plot(x='critic_score', y='general_sales', kind='scatter', alpha=0.3, figsize=(8,8), grid=True)
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –∫—Ä–∏—Ç–∏–∫–æ–≤ (XOne)')
plt.show()
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è XOne
xone_critic_cor = data_xone['critic_score'].corr(data_xone['general_sales'])
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (XOne): ', xone_critic_cor)


# 2 –ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è –ø–æ –æ—Ç–∑—ã–≤–∞–º –∫—Ä–∏—Ç–∏–∫–æ–≤ (2015-16 –≥–≥) –¥–ª—è 3DS
data_3ds.plot(x='critic_score', y='general_sales', kind='scatter', alpha=0.3, figsize=(8,8), grid=True)
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –∫—Ä–∏—Ç–∏–∫–æ–≤ (3DS)')
plt.show()
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è 3DS
critic_cor_3ds = data_3ds['critic_score'].corr(data_3ds['general_sales'])
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (3DS): ', critic_cor_3ds)


# 3 –ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è –ø–æ –æ—Ç–∑—ã–≤–∞–º –∫—Ä–∏—Ç–∏–∫–æ–≤ (2015-16 –≥–≥) –¥–ª—è WiiU
data_wiiu.plot(x='critic_score', y='general_sales', kind='scatter', alpha=0.3, figsize=(8,8), grid=True)
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –∫—Ä–∏—Ç–∏–∫–æ–≤ (WiiU)')
plt.show()
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è WiiU
wiiu_critic_cor = data_wiiu['critic_score'].corr(data_wiiu['general_sales'])
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (WiiU): ', wiiu_critic_cor)
```
