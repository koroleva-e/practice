```python
#зависит ли цена от общей площади
data.plot(x='last_price', y='total_area', kind='scatter', alpha=0.03);
plt.title('зависимость цены от общей площади')
plt.show()

print('Коэффициент корреляции: ', data['last_price'].corr(data['total_area']) )

#зависит ли цена от жилой площади
data.plot(x='last_price', y='living_area', kind='scatter', alpha=0.03);
plt.title('зависимость цены от жилой площади')
plt.show()

print('Коэффициент корреляции: ', data['last_price'].corr(data['living_area']) )

#зависит ли цена от площади кухни
data.plot(x='last_price', y='kitchen_area', kind='scatter', alpha=0.03);
plt.title('зависимость цены от площади кухни')
plt.show()

print('Коэффициент корреляции: ', data['last_price'].corr(data['kitchen_area']) )

#зависит ли цена от количества комнат
data.pivot_table(index='rooms',values='last_price').plot(y='last_price', kind='bar');
plt.title('зависимость цены от количества комнат')
plt.show()

print('Коэффициент корреляции: ', data['last_price'].corr(data['rooms']) )

#зависит ли цена от этажа, на котором расположена квартира (первый, последний, другой)
data.pivot_table(index='floor_type',values='last_price').plot(y='last_price', kind='bar');


#зависит ли цена от даты размещения (день недели, месяц, год)
data.plot(x='last_price', y='public_day', kind='scatter', alpha=0.03);
plt.title('зависимость цены от дня размещения')
plt.show()

print('Коэффициент корреляции: ', data['last_price'].corr(data['public_day']) )

data.plot(x='last_price', y='public_month', kind='scatter', alpha=0.03);
plt.title('зависимость цены от месяца размещения')
plt.show()

print('Коэффициент корреляции: ', data['last_price'].corr(data['public_month']) )

data.plot(x='last_price', y='public_year', kind='scatter', alpha=0.03);
plt.title('зависимость цены от года размещения')
plt.show()

print('Коэффициент корреляции: ', data['last_price'].corr(data['public_year']) )
```

Судя по графикам и коэффициенту коррелляции, цена зависит от:
- общей площади (cамый большой коэффициент)
- жилой площади
- площади кухни
- количества комнат
